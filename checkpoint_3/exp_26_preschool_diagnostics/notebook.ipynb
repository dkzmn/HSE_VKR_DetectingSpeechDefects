# Эксперимент 26: Preschool diagnostics (AI)

**Статья:** Диагностика речевых нарушений у дошкольников с помощью искусственного интеллекта 2024

**Ссылка:** [https://journal-iro23.ru/ru/все-выпуски/педагогическая-перспектива-2024-№-1/диагностика-речевых-нарушений/](https://journal-iro23.ru/ru/все-выпуски/педагогическая-перспектива-2024-№-1/диагностика-речевых-нарушений/)

**Краткое описание модели:** ASR-прокси акустические признаки + SVM для скрининга детской речи.

**Содержание статьи:** Статья релевантна возрастно и клинически: рассматриваются типы нарушений у дошкольников и практики ИИ-диагностики. Она задает важный прикладной контекст для скрининговых моделей. Эксперимент реализует компактный ML-подход, ориентированный на педиатрическую задачу.

```python
import sys
from pathlib import Path
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score, classification_report

exp_dir = Path().resolve()
sys.path.insert(0, str(exp_dir.parent))

from shared import config, data_utils
from shared.results_utils import save_result_csv
import librosa
```

## 1. Загрузка данных и разбиение

```python
paths_train, paths_val, paths_test, y_train, y_val, y_test, letters_train, letters_val, letters_test = data_utils.get_splits()
print(f"Train: {len(paths_train)}, Val: {len(paths_val)}, Test: {len(paths_test)}")
n_letters = letters_train.shape[1]
```

## 2. Извлечение признаков

```python

def extractor(path):
    # ASR-подобные признаки без транскрипта: энергия, ZCR, спектральные и pitch + MFCC stats
    y, sr = data_utils.load_audio(path, sr=config.TARGET_SR, max_sec=config.MAX_DURATION_SEC)
    rms = librosa.feature.rms(y=y, frame_length=config.WIN_LENGTH, hop_length=config.HOP_LENGTH)[0]
    zcr = librosa.feature.zero_crossing_rate(y, frame_length=config.WIN_LENGTH, hop_length=config.HOP_LENGTH)[0]
    cent = librosa.feature.spectral_centroid(y=y, sr=sr, hop_length=config.HOP_LENGTH)[0]
    roll = librosa.feature.spectral_rolloff(y=y, sr=sr, hop_length=config.HOP_LENGTH)[0]
    f0, _, _ = librosa.pyin(y, fmin=librosa.note_to_hz("C2"), fmax=librosa.note_to_hz("C7"), sr=sr, hop_length=config.HOP_LENGTH)
    f0 = np.nan_to_num(f0, nan=0.0)
    stats = []
    for arr in [rms, zcr, cent, roll, f0]:
        stats.extend([float(np.mean(arr)), float(np.std(arr)), float(np.percentile(arr, 75))])
    mfcc = data_utils.extract_mfcc_stats(path)
    return np.concatenate([np.array(stats, dtype=np.float32), mfcc]).astype(np.float32)
```

## 3. Обучение, подбор гиперпараметров и оценка

```python
X_train = data_utils.build_feature_matrix(paths_train, extractor, n_jobs=-1)
X_val   = data_utils.build_feature_matrix(paths_val, extractor, n_jobs=-1)
X_test  = data_utils.build_feature_matrix(paths_test, extractor, n_jobs=-1)
if letters_train.shape[1] > 0:
    X_train = np.hstack([X_train, letters_train])
    X_val   = np.hstack([X_val, letters_val])
    X_test  = np.hstack([X_test, letters_test])
print(f"Размер признакового вектора: {X_train.shape[1]}")

classes = np.unique(y_train)
weights = compute_class_weight("balanced", classes=classes, y=y_train)
class_weight = dict(zip(classes, weights))
base_model = SVC(kernel="rbf", class_weight=class_weight, probability=True, random_state=config.RANDOM_STATE)
param_grid = {
    "clf__C": [0.1, 1.0, 10.0], 
    "clf__gamma": ["scale", "auto"]
}

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", base_model),
])

grid = GridSearchCV(pipe, param_grid, cv=5, scoring="f1_macro", refit=True, n_jobs=-1, verbose=1)
t0 = time.perf_counter()
grid.fit(X_train, y_train)
train_time_sec = time.perf_counter() - t0
clf = grid.best_estimator_
print("Лучшие параметры:", grid.best_params_)

# График зависимости mean CV score от значений гиперпараметров
cv_results = grid.cv_results_
params = list(param_grid.keys())
fig, axes = plt.subplots(1, len(params), figsize=(5 * len(params), 4))
if len(params) == 1:
    axes = [axes]
for ax, pname in zip(axes, params):
    key = "param_" + pname
    vals = list(param_grid[pname])
    means = []
    for v in vals:
        mask = cv_results[key] == v
        means.append(np.mean(cv_results["mean_test_score"][mask]))
    ax.bar(range(len(vals)), means)
    ax.set_xticks(range(len(vals)))
    ax.set_xticklabels([str(v) for v in vals], rotation=45, ha="right")
    ax.set_title(pname.replace("clf__", ""))
    ax.set_ylabel("F1-macro CV")
plt.tight_layout(); plt.show()

y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:, 1]
accuracy = accuracy_score(y_test, y_pred)
f1_macro = f1_score(y_test, y_pred, average="macro")
f1_bad = f1_score(y_test, y_pred, average="binary", pos_label=config.CLASS_BAD)
roc_auc = roc_auc_score(y_test, y_proba)
precision_bad = precision_score(y_test, y_pred, zero_division=0, pos_label=config.CLASS_BAD)
recall_bad = recall_score(y_test, y_pred, zero_division=0, pos_label=config.CLASS_BAD)

print(classification_report(y_test, y_pred, target_names=config.CLASS_NAMES))
metrics_table = pd.DataFrame([{
    "accuracy": accuracy,
    "f1_macro": f1_macro,
    "f1_bad": f1_bad,
    "roc_auc": roc_auc,
    "precision_bad": precision_bad,
    "recall_bad": recall_bad,
}])
display(metrics_table)

save_result_csv(
    exp_dir=exp_dir,
    experiment_id="exp_26_preschool_diagnostics",
    experiment_name="Preschool diagnostics (AI)",
    model="SVM (RBF)",
    accuracy=accuracy,
    f1_macro=f1_macro,
    f1_bad=f1_bad,
    roc_auc=roc_auc,
    precision_bad=precision_bad,
    recall_bad=recall_bad,
    notes="asr_proxy/grid=" + str(grid.best_params_),
    num_params=None,
    train_time_sec=train_time_sec,
)
```
